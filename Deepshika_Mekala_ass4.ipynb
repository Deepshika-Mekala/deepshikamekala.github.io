```python
from yellowbrick.datasets import load_energy

X, y = load_energy()
print('X.shape={}, type(X)={}'.format(X.shape, type(X)))
print('y.shape={}, type(y)={}'.format(y.shape, type(y)))
```

```python
# Split your data into training and testing datasets (select random_state=0 and use default test_size)
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, random_state=0)
```

```python
# Implement scaling and/or encoding here if needed (2 marks for preprocessing properly or justifying why it isn't needed)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

column = ColumnTransformer(
    [("scaling", StandardScaler(), ['relative compactness', 'surface area', 'wall area', 'roof area','overall height','glazing area','glazing area distribution']),
     ("onehot", OneHotEncoder(sparse=False), ['orientation'])])

column.fit(X_train)
X_train_trans = column.transform(X_train)

X_test_trans = column.transform(X_test)
print('X_train_trans.shape={}, type(X_train_trans)={}'.format(X_train_trans.shape, type(X_train_trans)))
print('X_test_trans.shape={}, type(X_test_trans)={}'.format(X_test_trans.shape, type(X_test_trans)))
```

## Part 4: Implement MLP using scikit-learn (5 marks)

```python
# Test using default parameters (set max_iter=500 - for this assignment, don't worry about reaching convergence)
from sklearn.neural_network import MLPRegressor

from sklearn.datasets import fetch_california_housing
california = fetch_california_housing()

mlp = MLPRegressor(max_iter=500)
mlp.fit(X_train_trans, y_train)
rs = mlp.score(X_test_trans, y_test)
print("R2 score:", rs)
```

```python
# Test using two hidden layers with 100 nodes each
mlp2 = MLPRegressor(hidden_layer_sizes=(100, 100), max_iter=500)
mlp2.fit(X_train_trans, y_train)
rs = mlp2.score(X_test_trans, y_test)
print("R2 score:", rs)
```

```python
# Test using three hidden layers with 100 nodes each
mlp3 = MLPRegressor(hidden_layer_sizes=(100, 100, 100), max_iter=500)
mlp3.fit(X_train_trans, y_train)
rs = mlp3.score(X_test_trans, y_test)
print("R2 score:", rs)
```

## Part 5: Implement MLP using TensorFlow (7 marks)

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
```

Instead of scaling the data using a scikit-learn scaler, you can scale the data using a normalization layer.

```python
# Define normalization layer
normalization_layer = tf.keras.layers.experimental.preprocessing.Normalization()
normalization_layer.adapt(X_train)
```

Using `keras.Sequential`, implement an MLP with the same hidden layer setups as above:

```python
# One hidden layer with 100 nodes and the relu activation function
# Compile the model with loss='mean_absolute_error' and optimizer=tf.keras.optimizers.Adam(0.001)
# Fit the model using validation_split=0.2, verbose=0 and epochs=100
from tensorflow.keras.layers import Dense, Input, BatchNormalization

ten = tf.keras.models.Sequential([
    Input(shape=(X_train.shape[1],)),
    normalization_layer,
    tf.keras.layers.Dense(100, activation='relu'),
    BatchNormalization(),
    tf.keras.layers.Dense(1)
])

ten.compile(loss='mean_absolute_error', optimizer=tf.keras.optimizers.Adam(0.001))

out = ten.fit(X_train, y_train, validation_split=0.2, verbose=0, epochs=100)

score = ten.evaluate(X_test, y_test)
print("loss:", score)
```

```python
# Repeat with two hidden layers with 100 nodes each and the relu activation function
from tensorflow.keras.layers import Dense, Input, BatchNormalization

ten2 = tf.keras.models.Sequential([
    Input(shape=(X_train.shape[1],)),
    normalization_layer,
    tf.keras.layers.Dense(100, activation='relu'),
    BatchNormalization(),
    tf.keras.layers.Dense(100, activation='relu'),
    BatchNormalization(),
    tf.keras.layers.Dense(1)
])

ten2.compile(loss='mean_absolute_error', optimizer=tf.keras.optimizers.Adam(0.001))

out2 = ten2.fit(X_train, y_train, validation_split=0.2, epochs=100, verbose=0)

score = ten2.evaluate(X_test, y_test)
print("loss:", score)
```

```python
# Repeat with three hidden layers with 100 nodes each and the relu activation function
from tensorflow.keras.layers import Dense, Input, BatchNormalization
ten3 = tf.keras.models.Sequential([
    Input(shape=(X_train.shape[1],)),
    normalization_layer,
    tf.keras.layers.Dense(100, activation='relu'),
    BatchNormalization(),
    tf.keras.layers.Dense(100, activation='relu'),
    BatchNormalization(),
    tf.keras.layers.Dense(100, activation='relu'),
    BatchNormalization(),
    tf.keras.layers.Dense(1)
])

ten3.compile(loss='mean_absolute_error', optimizer=tf.keras.optimizers.Adam(0.001))

out3 = ten3.fit(X_train, y_train, validation_split=0.2, epochs=100, verbose=0)

score = ten3.evaluate(X_test, y_test)
print("loss:", score)
```

## Part 6: Compare the accuracy of both methods (7 marks)

For this part, calculate the mean absolute error for each model and print in a table using pandas

```python
# Calculate the MAE for the three scikit-learn tests
from sklearn.metrics import mean_absolute_error

sk1 = mlp.predict(X_test_trans)
m1 = mean_absolute_error(y_test, sk1)
print("Test1-MAE:", m1)

sk2 = mlp2.predict(X_test_trans)
m2 = mean_absolute_error(y_test, sk2)
print("Test2-MAE:", m2)

sk3 = mlp3.predict(X_test_trans)
m3 = mean_absolute_error(y_test, sk3)
print("Test3-MAE:", m3)
```

```python
# Calculate the MAE for the three tensorflow tests
tfp1 = ten.predict(X_test)
tm1 = mean_absolute_error(y_test, tfp1)
print("Test1-MAE:", tm1)

tfp2 = ten2.predict(X_test)
tm2 = mean_absolute_error(y_test, tfp2)
print("Test2-MAE:", tm2)

tfp3 = ten3.predict(X_test)
tm3 = mean_absolute_error(y_test, tfp3)
print("Test3-MAE:", tm3)
```
